```python
movies_df = pd.read_csv('movies.csv', usecols = ['movieId','title'])
rating_df = pd.read_csv('ratings.csv',usecols  =['userId', 'movieId','rating'])
```

```python
df = pd.merge(rating_df,movies_df,on = 'movieId')
df.head()
```

```python
df.isnull().sum()
```

```python
movie_ratings = pd.DataFrame(df.groupby('title')['userId'].count().reset_index())
```

```python
movie_ratings.head()
```

```python
movie_ratings.rename(columns={ 'userId': 'total_rating_count'}, inplace  = True)
```

```python
movie_ratings.head()
```

```python
df.head()
```

```python
df = pd.merge(df,movie_ratings,on = 'title')
```

```python
df.head()
```

```python
new_df = df[df['total_rating_count']>50]
```

```python
new_df.head()
```

```python
df.shape
```

```python
new_df.shape
```

```python
movie_features_df = new_df.pivot_table(index = 'title',columns = 'userId', values = 'rating').fillna(0)
movie_features_df.head()
```

```python
# converting pivot table into array matrx
from scipy.sparse import csr_matrix
movie_features_df_matrix = csr_matrix(movie_features_df.values)

from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine',algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
query_index = np.random.choice(movie_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
movie_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
